```python
%matplotlib inline
import numpy as np 
import pandas as pd
import networkx as nx 

from pylab import rcParams
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel(r"C:\Users\Ishi\Desktop\MCM Competition\TEEBDatabase.xlsx", sheet_name = 1)
data.isnull().sum()/len(data)
data
```

```python
services = []

for i in data.ESService:
    if i not in services:
        services.append(i)
services
```

```python
DG = nx.gn_graph(len(services))

for i in range(len(services)):
    DG.node[i]['name'] = services[i]
```

```python
DG = nx.gn_graph(len(services))

DG.add_edges_from([(0, 1)])
for i in range(len(services)):
    DG.node[i]['name'] = services[i]

labeldict = {}

for i in range(len(services)):
    labeldict[i] = DG.node[i]["name"]

plt.figure(figsize=(6,6))

nx.draw(DG, with_labels = True, labels = labeldict)
plt.axis("equal")
```
